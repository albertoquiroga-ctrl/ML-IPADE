## Loops

Loops let you repeat a block of code multiple times.

---

### `for` loop

Use a `for` loop when you know **how many times** you want to repeat something, or when you want to go through each item in a list.

```python
# Example 1: count from 1 to 5
for i in range(1, 6):
    print(i)

print("Done!")
```

```python
# Example 2: loop through a list of items
fruits = ["apple", "banana", "cherry"]

for fruit in fruits:
    print(f"I like {fruit}")
```

### `while` loop

Use a `while` loop when you want to keep repeating **as long as a condition is true**. Be careful — if the condition never becomes false, the loop runs forever!

```python
# Example 1: count down from 5 to 1
count = 5

while count > 0:
    print(count)
    count -= 1  # subtract 1 each time, otherwise the loop never ends!

print("Liftoff!")
```

```python
# Example 2: keep asking until the user guesses correctly
secret = "python"

guess = input("Guess the secret word: ")

while guess != secret:
    print("Wrong! Try again.")
    guess = input("Guess the secret word: ")

print("Correct! You got it.")
```

## Conditionals

Conditionals let you run code **only when a condition is true**.

---

### `if` / `elif` / `else`

- `if` — runs the block if the condition is **true**
- `elif` — checks another condition if the previous one was false
- `else` — runs if **none** of the above conditions were true

```python
# Example 1: check if a number is positive, negative, or zero
number = 7

if number > 0:
    print("Positive")
elif number < 0:
    print("Negative")
else:
    print("Zero")
```

```python
# Example 2: assign a letter grade based on a score
score = 85

if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
elif score >= 70:
    grade = "C"
else:
    grade = "F"

print(f"Score: {score} → Grade: {grade}")
```
